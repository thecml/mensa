```python
import numpy as np
import pandas as pd
import config as cfg

matplotlib_style = 'default'
import matplotlib.pyplot as plt; plt.style.use(matplotlib_style)
plt.rcParams.update({'axes.labelsize': 'medium',
                     'axes.titlesize': 'medium',
                     'font.size': 14.0,
                     'text.usetex': True,
                     'text.latex.preamble': r'\usepackage{amsfonts} \usepackage{bm}'})

from utility.plot import make_bar_plot, TFColor

filename = f"{cfg.RESULTS_DIR}/synthetic_cr_theta.csv"
df = pd.read_csv(filename)
df = df.round(3)

# Add Epoch
df['Epoch'] = df.groupby('KTau').cumcount() + 1
```

```python
unique_k_tau_values = df['KTau'].unique()

# Plotting
for k_tau in unique_k_tau_values:
    subset_df = df[df['KTau'] == k_tau]
    plt.figure(figsize=(6,4))
    
    for value in ['KTau_1', 'KTau_2', 'KTau_3', 'KTau_4']:
        plt.plot(subset_df['Epoch'], subset_df[value], label=value)
    
    plt.xlabel("Epoch")
    plt.ylabel("Values")
    plt.title(f"Values as a Function of Epochs for KTau = {k_tau}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
